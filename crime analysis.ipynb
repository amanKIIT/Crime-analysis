```python
#Q2.Show bar plot of top 20 district with maximum rape cases
```

```python
piv_rape=pd.pivot_table(df, values ='Rape', index =['DISTRICT'], aggfunc = sum)
```

```python
top_rapes_by_dist = piv_rape.sort_values(by=['Rape'], ascending=False)
```

```python
rape_top_20_dist = top_rapes_by_dist.head(20)
rape_top_20 = rape_top_20_dist.drop('TOTAL')
rape_top_20
```

```python
rape_top_20.plot.bar(color = 'brown')
```

```python
#Q3.Show bar plot of top 20 State having maximum kidnapping and abduction
```

```python
piv_kid = pd.pivot_table(df, values ='Kidnapping and Abduction', index =['STATE/UT'], aggfunc = sum)
```

```python
top_kid_by_state = piv_kid.sort_values(by=['Kidnapping and Abduction'], ascending=False)
kid_top_20_state = top_kid_by_state.head(20)
kid_top_20_state
```

```python
kid_top_20_state.plot.bar(color = 'orange')
```
